```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\ML\exp.csv")
if df.shape[1] >= 2:
    plt.figure(figsize=(6,5))
    for i in range(len(df)):
        plt.plot([0,1], [df.iloc[i,0], df.iloc[i,1]], marker="o")
    plt.xticks([0,1], [df.columns[0], df.columns[1]])
    plt.title("Slope Chart - First 2 Columns")
    plt.ylabel("Death Rate")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r"C:\ML\exp.csv")

values = df.sum(numeric_only=True).sort_values(ascending=False)
cum_percent = values.cumsum() / values.sum() * 100

fig, ax = plt.subplots()

ax.bar(values.index, values)
ax.set_ylabel("Frequency")

ax2 = ax.twinx()
ax2.plot(values.index, cum_percent, color="red", marker="D")
ax2.set_ylabel("Cumulative Percentage")

plt.title("Pareto Chart")
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv(r"C:\ML\exp.csv")

numeric_df = df.select_dtypes(include=np.number)
fig = px.bar_polar(r=numeric_df.iloc[0],theta=numeric_df.columns,template="plotly_dark")
fig.update_layout(title="Radial Bar Chart")
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\ML\exp.csv")
categories=list(df.columns)
values=df.iloc[0].values.flatten().tolist()
values+=values[:1]
angles=np.linspace(0,2*np.pi,len(categories),endpoint=False).tolist()
angles+=angles[:1]
plt.figure(figsize=(6,6))
ax=plt.subplot(111,polar=True)
ax.plot(angles,values,marker="o")
ax.fill(angles,values,alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("radar chart")
plt.show()
```
